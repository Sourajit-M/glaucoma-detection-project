```python
# ── Robust project-root resolution ───────────────────────────────────
# Works regardless of where you launched Jupyter from.
# Walks up from this notebook's location until it finds config.py.
import sys
from pathlib import Path

def _find_project_root(start: Path, marker: str = 'config.py') -> Path:
    """Walk up the directory tree until we find the project root marker."""
    for parent in [start, *start.parents]:
        if (parent / marker).exists():
            return parent
    raise FileNotFoundError(
        f"Could not find '{marker}' in any parent of {start}.\n"
        "Make sure you are running this notebook from inside the project folder."
    )

PROJECT_ROOT = _find_project_root(Path().resolve())

if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

print(f'Project root : {PROJECT_ROOT}')
print(f'sys.path[0]  : {sys.path[0]}')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2

from config import FIGURES_DIR, DEVICE
from data.dataset_loader import (
    load_acrima,
    load_rimone,
    load_airogs,
    load_drishti_segmentation,
    load_all_datasets,
    print_dataset_summary,
)
from data.preprocessing import preprocess_image, verify_dataset

plt.rcParams['figure.dpi'] = 120
print(f'Imports OK  |  Device: {DEVICE}')
```

## 1.1 Load each dataset individually

```python
acrima_df  = load_acrima()
rimone_df  = load_rimone()
airogs_df  = load_airogs(max_samples=2000)   # cap train split; val+test loaded in full
drishti_df = load_drishti_segmentation()
```

## 1.2 Combined classification DataFrame

```python
df = load_all_datasets(include=['ACRIMA', 'RIMONE', 'AIROGS'], airogs_max_samples=2000)
print_dataset_summary(df)
df.head()
```

## 1.3 Class distribution per dataset

```python
fig, axes = plt.subplots(1, 3, figsize=(14, 4))
colours = ['#3B8BD4', '#E8593C']

for ax, (name, sub_df) in zip(axes, df.groupby('dataset')):
    counts = sub_df['label'].value_counts().sort_index()
    bars = ax.bar(['Normal', 'Glaucoma'], counts.values, color=colours, width=0.4)
    ax.set_title(name, fontsize=13, fontweight='bold')
    ax.set_ylabel('Images')
    for bar, val in zip(bars, counts.values):
        ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 2,
                str(val), ha='center', va='bottom', fontsize=10)
    ratio = counts.max() / counts.min()
    ax.set_xlabel(f'Imbalance ratio: {ratio:.2f}x', fontsize=9)

plt.suptitle('Class Distribution per Dataset', fontsize=14)
plt.tight_layout()
plt.savefig(FIGURES_DIR / 'class_distribution.png', bbox_inches='tight')
plt.show()
print(f'Saved → {FIGURES_DIR / "class_distribution.png"}')
```

## 1.4 Sample images — preprocessed (CLAHE + circular mask)

```python
samples_per_class = 2
datasets_to_show  = ['ACRIMA', 'RIMONE', 'AIROGS']

fig, axes = plt.subplots(
    len(datasets_to_show), samples_per_class * 2,
    figsize=(16, 4 * len(datasets_to_show))
)

for row_idx, ds_name in enumerate(datasets_to_show):
    sub = df[df['dataset'] == ds_name]
    samples = pd.concat([
        sub[sub['label'] == 0].sample(samples_per_class, random_state=42),
        sub[sub['label'] == 1].sample(samples_per_class, random_state=42),
    ])
    for col_idx, (_, row) in enumerate(samples.iterrows()):
        ax = axes[row_idx][col_idx]
        img = preprocess_image(row['image_path'], use_clahe=True, use_circular_mask=True)
        ax.imshow(img)
        label_str = 'Glaucoma' if row['label'] == 1 else 'Normal'
        colour    = '#E8593C'  if row['label'] == 1 else '#3B8BD4'
        ax.set_title(f'{ds_name} — {label_str}', color=colour, fontsize=9)
        ax.axis('off')

plt.suptitle('Sample Images (CLAHE + circular mask applied)', fontsize=13)
plt.tight_layout()
plt.savefig(FIGURES_DIR / 'sample_images.png', bbox_inches='tight')
plt.show()
print(f'Saved → {FIGURES_DIR / "sample_images.png"}')
```

## 1.5 DRISHTI-GS1 — segmentation mask overlay

```python
valid = drishti_df.dropna(subset=['od_mask_path', 'cup_mask_path']).head(4)

if len(valid) == 0:
    print('No DRISHTI samples with both masks found.')
    print('Run debug_drishti_gt() from dataset_loader to inspect your GT/ folder.')
else:
    fig, axes = plt.subplots(3, len(valid), figsize=(4 * len(valid), 10))
    if len(valid) == 1:
        axes = axes.reshape(3, 1)   # keep 2-D indexing consistent

    for col, (_, row) in enumerate(valid.iterrows()):
        img     = cv2.cvtColor(cv2.imread(row['image_path']), cv2.COLOR_BGR2RGB)
        od_mask = cv2.imread(str(row['od_mask_path']),  cv2.IMREAD_GRAYSCALE)
        cu_mask = cv2.imread(str(row['cup_mask_path']), cv2.IMREAD_GRAYSCALE)

        overlay = img.copy()
        if od_mask is not None: overlay[od_mask > 127] = [0,   200,  0]    # OD  → green
        if cu_mask is not None: overlay[cu_mask > 127] = [200,  50, 50]    # cup → red

        for ax, im, title in zip(
            axes[:, col],
            [img, od_mask, overlay],
            ['Fundus', 'OD mask', 'Overlay (green=disc, red=cup)']
        ):
            ax.imshow(im, cmap='gray' if title == 'OD mask' else None)
            ax.set_title(title, fontsize=8)
            ax.axis('off')

    plt.suptitle('DRISHTI-GS1 — Segmentation Ground Truth', fontsize=13)
    plt.tight_layout()
    plt.savefig(FIGURES_DIR / 'drishti_segmentation_samples.png', bbox_inches='tight')
    plt.show()
    print(f'Saved → {FIGURES_DIR / "drishti_segmentation_samples.png"}')
```

## 1.6 Per-dataset image statistics (mean & std RGB)

```python
def compute_image_stats(df, n_sample=200):
    sample = df.sample(min(n_sample, len(df)), random_state=42)
    means, stds = [], []
    for _, row in sample.iterrows():
        img = cv2.imread(str(row['image_path']))
        if img is None:
            continue
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img = cv2.resize(img, (224, 224)).astype(np.float32) / 255.0
        means.append(img.mean(axis=(0, 1)))
        stds.append(img.std(axis=(0, 1)))
    return np.mean(means, axis=0), np.mean(stds, axis=0)

print(f'{'Dataset':<12}  {'Mean (R,G,B)':>22}   {'Std (R,G,B)':>22}')
print('─' * 62)
for ds_name in df['dataset'].unique():
    sub = df[df['dataset'] == ds_name]
    mean, std = compute_image_stats(sub)
    print(f'{ds_name:<12}  {str(mean.round(3)):>22}   {str(std.round(3)):>22}')
```
